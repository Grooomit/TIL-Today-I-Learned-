#### 대회 평가지표
월간 데이콘 8회 대회의 평가지표는 AUC 입니다. 강의 시간에 배운 평가 지표로는 정확도, 정밀도, 재현율, F1_Score 등이 있었는데 이러한 평가 지표들은 라벨 값을 이용해 계산하게 됩니다. <br>
하지만 AUC는 라벨 값이 아닌 모델이 해당 클래스를 예측한 확률을 이용해 평가하는 지표입니다. <br>
따라서 결과를 만드실 때, predict() 함수가 아닌 predict_proba() 라는 함수를 사용해야합니다. 제출 파일을 생성하는 코드는 남겨드리니 확률 값만 넣어주시면 됩니다.<br>
남겨드린 참고 링크를 확인해보시고, 이해가 가지 않는 부분은 질문 부탁드리겠습니다.

#### Feature Engineering

```python
dataset = [train, test]

for data in dataset:
  data['T'] = data['QcA'] - data['QfA'] + data['QoA'] - data['QrA'] + data['QsA']
  data['V'] = data['QbA'] - data['QeA'] + data['QhA'] + data['QjA'] + data['QmA'] - data['QqA']
  data['M'] = - data['QkA']
```

```python
# 주어진 데이터 그대로 문항간 Spearman correlation을 계산
questions = [i for i in list(ascii_lowercase)[:20]]
answers = [('Q'+i+'A') for i in questions]
```

```python
# 이미 알고 있는 -부호 문항을 reverse 처리하기
flipping_columns = ["QeA", "QfA", "QkA", "QqA", "QrA"]

for data in dataset:
  for flip in flipping_columns: 
      data[flip] = 6 - data[flip]
```

```python
# -부호인 Secret 문항을 reverse 처리하기
flipping_secret_columns = ["QaA", "QdA", "QgA", "QiA", "QnA"]

for data in dataset:
  for flip in flipping_secret_columns: 
      data[flip] = 6 - data[flip]
```

```python
# 마키아벨리니즘 테스트 스코어를 나타내는 Mach_score column을 추가
train['Mach_score'] = train[answers].mean(axis = 1)
test['Mach_score'] = test[answers].mean(axis = 1)
```

```python
# 응답시간 총합 / 10
for data in dataset:
  data['delay'] = data[[('Q'+i+'E') for i in questions]].sum(axis=1)
  data['delay'] = data['delay'] ** (1/10)
```

```python
delay = [('Q'+i+'E') for i in questions]

for data in dataset:
  data.drop(delay, axis=1, inplace=True)
```

```python
# 불필요한 컬럼 삭제
wr_list = [('wr_0'+str(i)) for i in range(1,10)]
wr_list.extend([('wr_'+str(i)) for i in range(10,14)])
wr_no_need = [i for i in wr_list if i not in ['wr_01', 'wr_03', 'wr_06', 'wr_09', 'wr_11']]

for data in dataset:
  data.drop(wr_no_need, axis=1, inplace = True)
  data.drop('index', axis=1, inplace = True)
```

- round : 점수 반올림으로 표시할 자리를 적습니다.

#### 6. Tune Models
> tune_model(estimator = None, fold = 10, round = 4, n_iter = 10, custom_grid = None, optimize = ‘Accuracy’, choose_better = False, verbose = True)

- estimator : 사용할 모델을 입력합니다.
- fold : K-fold의 수로, 최소 2 이상의 숫자를 입력합니다.
- round : 점수 반올림으로 표시할 자리를 적습니다.
- n_iter : Random Grid Search를 한 회차당 반복할 회수입니다.
- custom_grid : 직접 파라미터의 범위를 조정할 수 있습니다.
- optimize : 파라미터 튜닝 과정에서 어떤 점수를 따라갈 것인지 선택 가능합니다. 'Accuracy','AUC','Recall','Precision', 'F1'
- choose_better : True로 설정하면 성능이 높아지지 않을 경우 tuning을 하지 않은 모델을 반환합니다.
- verbose : True로 설정하면 진행중인 상황이 나타납니다.
