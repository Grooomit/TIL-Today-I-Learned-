#### 대회 평가지표
월간 데이콘 8회 대회의 평가지표는 AUC 입니다. 강의 시간에 배운 평가 지표로는 정확도, 정밀도, 재현율, F1_Score 등이 있었는데 이러한 평가 지표들은 라벨 값을 이용해 계산하게 됩니다. <br>
하지만 AUC는 라벨 값이 아닌 모델이 해당 클래스를 예측한 확률을 이용해 평가하는 지표입니다. <br>
따라서 결과를 만드실 때, predict() 함수가 아닌 predict_proba() 라는 함수를 사용해야합니다. 제출 파일을 생성하는 코드는 남겨드리니 확률 값만 넣어주시면 됩니다.<br>
남겨드린 참고 링크를 확인해보시고, 이해가 가지 않는 부분은 질문 부탁드리겠습니다.

#### 1-2. 스케일링

```python
x_train_mean = np.mean(x_train[num_columns], axis=0)
x_train_std  = np.std(x_train[num_columns], axis=0)

# Numpy 브로드캐스팅 기능을 활용해 x_train의 평균과 표준편차로 x_train, x_valid의 스케일링을 진행해줍니다.
x_train.loc[:, num_columns] = (x_train[num_columns] - x_train_mean) / (x_train_std + 1e-10)
x_valid.loc[:, num_columns] = (x_valid[num_columns] - x_train_mean) / (x_train_std + 1e-10)
x_test.loc[:, num_columns]  = (x_test[num_columns]  - x_train_mean) / (x_train_std + 1e-10)
```

- 네가지 분류 모델의 성능을 비교해보세요. <br>
- 해당 대회의 평가지표는 AUC 이므로 "Scikit-learn roc_auc_score"라고 검색하여 해당 평가 지표를 사용해 Validation 성능을 비교합니다.<br> 그 중 가장 좋은 성능의 모델을 선택해 Test 데이터의 결과를 예측합니다. AUC는 높을 수록 좋은 모델입니다.
- 모델 파라미터는 지난 수업에서 제가 추천해드렸던 파라미터의 값을 실제로 변경해보고, 모델 튜닝도 시도해보시는 것을 추천드립니다. <br> 강의 시간에도 답변드렸지만, 실제로 파라미터 튜닝은 컴퓨터가 최적화하도록 하는 것이 효율적이니 한번 해보는 정도로만 해보셔도 괜찮습니다. (random_state 값은 고정시키는 것을 강력히 추천드립니다.)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
```

```python
# 모델 초기화
models = {
    'Logistic Regression': LogisticRegression(C=1.0, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
    'XGBoost': XGBClassifier(n_estimators=100, max_depth=10, learning_rate=0.1, random_state=42),
    'LightGBM': LGBMClassifier(n_estimators=100, max_depth=10, learning_rate=0.01, random_state=42)
}
```

```python
# # 교차 검증 수행
# n_folds = 5
# kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
```

```python
auc_scores = {}

for name, model in models.items():

    model.fit(x_train, y_train)
    pred = model.predict_proba(x_valid)[:, 1]
    auc = roc_auc_score(y_valid, pred)
    auc_scores[name] = auc
    
    print(f"{name} AUC: {auc:.4f}")

# 가장 높은 AUC 점수를 가진 모델을 선택합니다.
best_model_name = max(auc_scores, key=auc_scores.get)
best_model = models[best_model_name]
print("Best Model: ", best_model_name)

# 선택된 모델로 Test 데이터의 결과를 예측합니다.
pred = best_model.predict_proba(x_test)
print("Final Test Prediction: ", pred[:, 1] + 1)
```
